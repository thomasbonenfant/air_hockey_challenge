```python
# Extract data
J = np.array(data["performance_rho"])
lambdas = np.array(data["lambda_history"])
if CVAR:
    etas = np.array(data["eta_history"])
const = np.array(data["constraint_history"])
lagrangian = np.array(data["lagrangian"])

ite = len(J)
```

```python
# Plot J
plt.clf()
plt.figure(figsize=(20,8))
x = np.arange(ite)
plt.plot(x, J, "-", label="J")
plt.grid()
plt.legend(loc="best")
plt.title("Performance index J")
plt.xlabel("Iteration")
plt.ylabel("J value")
plt.show()
```

```python
# Plot L
plt.clf()
plt.figure(figsize=(20,8))
x = np.arange(ite)
plt.plot(x, lagrangian, "-", label="L(...)")
plt.grid()
plt.legend(loc="best")
plt.title("Lagrangian")
plt.xlabel("Iteration")
plt.ylabel("L value")
plt.show()
```

```python
# Plot Eta and Lambda
plt.clf()
plt.figure(figsize=(20,8))
x = np.arange(ite)
for i in range(n_constraints):
    if CVAR:
        plt.plot(x, etas[:,i], "-", label=f"Eta {i}")
    plt.plot(x, lambdas[:,i], "-", label=f"Lambda {i}")
plt.grid()
plt.legend(loc="best")
if CVAR:
    plt.title("Lambda and Eta")
else:
    plt.title("Lambda")
plt.xlabel("Iteration")
plt.ylabel("Eta and Lambda")
plt.show()
```

```python
# Plot Constraints
plt.clf()
plt.figure(figsize=(20,8))
x = np.arange(ite)
for i in range(n_constraints):
    plt.plot(x, const[:,i], "-", label=f"Constraint {i}")
plt.plot(x, np.zeros(ite), "-", color="black")
plt.grid()
plt.legend(loc="best")
plt.title("Constraints")
plt.xlabel("Iteration")
plt.ylabel("Constraint values")
plt.show()
```
